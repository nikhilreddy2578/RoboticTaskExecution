# Generating output with Gemini

```python
def gemini_task_tree(dish_name, ingredients):
    prompt = f"""
    Input: ● Dish name and preferred ingredients, ● Available ingredients and utensils in the kitchen (including their states - e.g.,
chopped onions, clean pot)
    Output: Task trees for each dish in the cafeteria menu.

    Generate a task tree for cooking the dish "{dish_name}" using the following ingredients:
    {", ".join(ingredients)}

    The task tree should be in JSON format with the following structure (instead of double curly braces use one):
    [
        {{
            "input_nodes": [
                {{
                    "label": "cutting board",
                    "states": [
                        "empty"
                    ],
                    "ingredients": [],
                    "container": null
                }},
                {{
                    "label": "tomato",
                    "states": [
                        "whole"
                    ],
                    "ingredients": [],
                    "container": null
                }}
            ],
            "motion_node": "pick-and-place",
            "output_nodes": [
                {{
                    "label": "cutting board",
                    "states": [],
                    "ingredients": [
                        "tomato"
                    ],
                    "container": null
                }},
                {{
                    "label": "tomato",
                    "states": [
                        "whole"
                    ],
                    "ingredients": [],
                    "container": "cutting board"
                }}
            ]
        }},
        ...
    ]
    Your response should contain only JSON  (so I can use json.loads())

    Using kitchen items:
        {kitchen}

    **Goal:**  if can_make_dish  else 'Unfortunately, it looks like you don\'t have all the ingredients to make {dish_name}. Generate a detailed task tree for cooking {dish_name}.'

    Please refer to the provided examples and structure within the prompt text for the desired format of the task tree.

    Use your knowledge and creativity to generate a similar task tree for {dish_name}.
    """

    response = model.generate_content(prompt)
    start_index = response.candidates[0].content.parts[0].text.find('{')
    end_index = response.candidates[0].content.parts[0].text.rfind('}') + 1
    json_string = response.candidates[0].content.parts[0].text[start_index:end_index]
    return json_string
```
